```python
import matplotlib.pyplot as plt


def mamas_methode(jahre: int, kapital: float, zinssatz: float):
    res = []
    jahrel = []
    
    for i in range(jahre):
        kapital = round((kapital / 100) * (100 + zinssatz), 2)
        
        res.append(kapital)
        jahrel.append(i)
    
    return [res, jahrel]


def papas_methode(jahre: int, kapital: float, zinssatz: float):
    res = []
    jahrel = []
    
    gesamt = kapital

    for i in range(jahre):
        wert = round((kapital / 100) * (100 + zinssatz), 2) - kapital
        gesamt += wert
        
        res.append(gesamt)
        jahrel.append(i)

    return [res, jahrel]

res = papas_methode(10, 10000, 5.2)
res2 = mamas_methode(10, 10000, 5.2)

fig, ax = plt.subplots()

plt.title("Beide Methoden im Vergleich")
plt.xlabel("Jahre")
plt.ylabel("Vermögen in €")

ax.plot(res[1], res[0], color='b', label='Papas Methode')
ax.plot(res2[1], res2[0], color='r', label='Mamas Methode')

leg = ax.legend();

fig.show()
```
